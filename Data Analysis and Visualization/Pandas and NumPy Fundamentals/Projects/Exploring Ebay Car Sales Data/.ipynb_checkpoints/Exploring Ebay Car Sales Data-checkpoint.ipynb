# 6. Dealing with Incorrect Registration Year Data

We can see that there are many registration years which are simply not reasonable. There are many registration years that are from far in the future and others that are frim far in the past before cars were invented. We can remove any rows with a registration year that is outside of a certain range.

```python
autos=autos[(autos["registration_year"]<=2021) & (autos["registration_year"]>1900)]
```

We can see that the most common registration month is "0". Since there is no 0th month, it is possible that 0 was put in as typo. Though the frequency of this likely suggests that 0 was placed for listings without a stated registration month. Since these make up early 10% of our dataset, and knowing the exact registration month is not crucial to us, we shall leave this alone. If, however, we were using this data for something that did require the exact registration month, we could do so as follows:

```python
autos_fixed_registration_month=autos[autos["registration_month"]>0]
```

```python
autos["registration_year"].value_counts(normalize=True)
```

We can see that the most common registration years were for recent years, but not too recent. About 10-15 years before the data was crawled. This makes sense as these are used cars which people often keep for that long before selling.

Unsurprisingly, the least common registration years were of cars that were 80+ years old. This is likely due to the fact that most cars from back then have broken down and have been scrapped and are not functional anymore.

# 7. Exploring Price by Brand

```python
brands = autos['brand'].value_counts(normalize = True)
all_brands = brands[brands > 0.0].index


brands_mean_price = {}

for brands in all_brands:
    brand_only = autos[autos['brand'] == brands]
    mean_price = brand_only['price'].mean()
    brands_mean_price[brands] = int(mean_price)
    
brands_mean_price.values
```

Above we can see that the most expensive brand is Range Rover. Next is "sonstige_autos", however that is simply the German term for "other cars", meaning other brands. Looking at this data, we can see that this "brand" consists of less common cars.
