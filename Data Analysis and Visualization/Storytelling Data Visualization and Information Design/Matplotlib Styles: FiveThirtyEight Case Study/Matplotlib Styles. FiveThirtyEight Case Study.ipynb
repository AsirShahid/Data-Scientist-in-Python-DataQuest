# 2. Wine Quality Dataset

```python
import pandas as pd
red_wine = pd.read_csv('winequality-red.csv', sep=';')
red_corr = red_wine.corr()['quality'][:-1]

print(red_corr)

white_wine=pd.read_csv("winequality-white.csv",sep=";")
white_corr=white_wine.corr()["quality"][:-1]

print(white_corr)

# For White Wine, the strongest correlation is with alcohol level, same as Red Wine. However, after that the 
# strongest correlation is with density and chlorides
```

# 3. FiveThirtyEight Style

```python
# Initial Code

style.use('fivethirtyeight')
fig, ax = plt.subplots(figsize=(9, 5))
ax.barh(white_corr.index, white_corr, left=2)
ax.barh(red_corr.index, red_corr)

```
